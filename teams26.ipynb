```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

df =pd.read_csv('teams.csv')
df.head()
```

```python
average = df.groupby('team')['avg_age'].sum().sort_values(ascending = False).head(10)
average
```

```python
average.plot(x=average.index,y=average.values,kind= 'bar')
```

```python
average.plot(y=average.index,x=average.values,kind='pie')
```

```python
average = pd.DataFrame({
    'team': ['Argentina', 'Brazil', 'Germany', 'Spain'],
    'avg_age': [29.4, 29.8, 28.0, 27.3]
})

# Plot a pie chart of average ages
plt.figure(figsize=(6,6))
plt.pie(
    average['avg_age'], 
    labels=average['team'], 
    autopct='%1.1f%%',   # <-- adds percentage values
    startangle=90
)
plt.title("Average Age of Teams")
plt.show()
```
