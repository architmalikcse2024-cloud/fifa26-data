```python
matches = pd.read_csv("matches.csv")
```

```python
matches['attendance'] = matches['attendance'].str.replace(',', '').astype(int)

# Plot attendance as a bar chart
plt.figure(figsize=(10,5))
plt.bar(matches['home_team'], matches['attendance'])
plt.xticks(rotation=90)
plt.title("Attendance per Match (by Home Team)")
plt.xlabel("Home Team")
plt.ylabel("Attendance")
plt.show()
```

```python
goals = df.assign(total_goals=df['home_score'] + df['away_score']).sort_values('total_goals', ascending=False).head(3)
```

```python
df['draws']=df['home_score'] == df['away_score']
df['draws'] = df['draws'].astype(int)

df[['home_team','away_team','draws']]
```

```python
teamdraw = df.groupby('home_team')['draws'].sum()
```
